```python
from collections import Counter
# Counter 문자개수를 딕셔너리로 반환
def solution(n, control):
    counts = Counter(control)
    return n + 1*(counts['w']-counts['s']) + 10*(counts['d']-counts['a'])
```

```python
print(solution(n,control))
```

```python
# zip함수로 dict 만들기
def solution(n, control):
    key = dict(zip(['w','s','d','a'], [1,-1,10,-10]))
    return n + sum([key[c] for c in control])
```

```python
# dict만들고 반복문
def solution(n, control):
    answer = n
    c = { 'w':1, 's':-1, 'd':10, 'a':-10}
    for i in control:
        answer += c[i]
    return answer
```
