```python
## 등차수열을 만들어서 풀기
def solution(a, d, included):
    answer = 0
    n = len(included)
    arr = [] # 등차수열
    true_list = []
    
    for i in range(n):
        arr.append(a + d*i)
        if included[i]:
            true_list.append(i)
    for x in true_list:
       answer += arr[x]
    return answer
```

```python
# 등차수열 안만들고 풀이
def solution(a, d, included):
    answer = 0
    true_list = []
    for i in range(len(included)):
        if included[i]:
            true_list.append(a + d*i) # 0,3,4 => 등차수열이 있다고 가정하고 0,3,4항의값만 더하면가능
    answer = sum(true_list)
    return answer
```
