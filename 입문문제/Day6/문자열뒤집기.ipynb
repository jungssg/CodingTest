## 문제 설명
문자열 my_string이 매개변수로 주어집니다. my_string을 거꾸로 뒤집은 문자열을 return하도록 solution 함수를 완성해주세요.  

- 제한사항  
1 ≤ my_string의 길이 ≤ 1,000

```python
def solution(my_string):
    return my_string
```

```python
def solution(my_string):
    my_string = list(my_string)
    my_string.reverse()
    return ''.join(my_string)
```

```python
print(solution('abc'))
```
