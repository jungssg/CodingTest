## 문제 설명
정수 n이 주어질 때, n이하의 짝수를 모두 더한 값을 return 하도록 solution 함수를 작성해주세요.

- 제한사항
0 < n ≤ 1000

```python
def solution(n):
    sum_n = 0
    for i in range(1,n+1):
        if i%2==0:
            sum_n += i
    answer = sum_n
    return answer
```

```python
def solution(n):
    return sum(i for i in range(0,n+1,2))
```

range는 이미 iterable 객체이기 때문에 list comprehension으로 다시 풀어 쓸 필요 없이 바로 sum 함수에 넣을 수 있습니다.

```python
def solution(n):
    return sum(range(0,n+1,2))
```
