1 ≤ my_string의 길이 ≤ 1,000  
my_string은 영어 대문자와 소문자로만 구성되어 있습니다.

"cccCCC"	"CCCccc"  
"abCdEfghIJ"	"ABcDeFGHij"

소문자는 대문자로 대문자는 소문자로 바꾼 "ABcDeFGHij"를 return합니다.
