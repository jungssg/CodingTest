```python
from collections import Counter
def solution(array):
    counts = Counter(array).most_common(2)
    if len(counts)==1: ## array=[1] 일때
        return counts[0][0]
    if counts[0][1] == counts[1][1]:
        return -1
    return counts[0][0]
```
