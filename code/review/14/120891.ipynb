## 문제 풀이

```python
def solution(order):
    return sum([str(order).count(s) for s in '369'])

solution(29423)
```

```python
def solution(order):
    clap = 0
    for s in str(order):
        if s in '369':
            clap+=1
    return clap

solution(29423)
```
