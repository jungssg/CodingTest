## 가까운 수


https://school.programmers.co.kr/learn/courses/30/lessons/120890

## 문제 풀이

```python
def solution(array, n):
    array = sorted(array)
    idx_diff = dict(enumerate(map(lambda a: abs(a-n), array)))
    idx = min(idx_diff, key=idx_diff.get)
    return array[idx]

solution([1, -1, 2], 0)
```
