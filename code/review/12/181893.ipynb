## 문제 풀이

```python
def solution(arr, query):
    for n, idx in enumerate(query):
        if n%2==0:
            arr = arr[:idx+1]
        else:
            arr = arr[idx:]
    return arr

solution([0, 1, 2, 3, 4, 5], [4, 1, 2])
```
