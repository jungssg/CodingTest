## 문제 풀이

```python
def solution(arr):
    start, end = 0,0
    for i in range(len(arr)):
        if arr[i]==2:
            start=i
            break
    for i in range(len(arr)):
        if arr[-i-1]==2:
            end=-i-1
            end = None if end==-1 else end+1
            break
    return arr[start:end] or [-1]

solution([2, 2, 2, 2])
```
