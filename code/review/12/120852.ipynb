## 문제 풀이

```python
def solution(n):
    i = 1
    factor = set()
    while n>1:
        i+=1
        while n%i==0:
            factor.add(i)
            n//=i
    factor = list(factor)
    return sorted(factor)

solution(100)
```

## 오답 풀이

```python
def solution(n):
    factor = []
    for i in range(2,n+1):
        if n==1:
            break
        if n%i==0:
            factor.append(i)
            n//=i
    return factor

solution(100)
# solution(97)
```

```python
def solution(n):
    factor = []
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            factor.append(i)
            n//=i
    return factor or [n]

solution(100)
# solution(97)
```
