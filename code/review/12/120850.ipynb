## 문제 풀이

```python
def solution(my_string):
    answer = []
    for num in my_string :
        if num in map(str,range(10)):
            answer.append(int(num))
    return sorted(answer)

solution("hi12392")
```

```python
def solution(my_string):
    my_string = sorted(my_string)
    for idx, num in enumerate(my_string) :
        if num not in map(str,range(10)):
            return my_string[:idx]

solution("hi12392")
```

```python
def solution(my_string):
    num_arr = []
    for num in range(10):
        cnt = sorted(my_string).count(str(num))
        if cnt:
            num_arr.extend([num]*cnt)
    return num_arr

solution("hi12392")
```
