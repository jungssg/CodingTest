## 문자 개수 세기

## 문제 풀이

```python
def solution(my_string):
    case_cnt = []
    for upper in range(ord('A'),ord('Z')+1):
        case_cnt.append(my_string.count(chr(upper)))
    for lower in range(ord('a'),ord('z')+1):
        case_cnt.append(my_string.count(chr(lower)))
    return case_cnt

solution("Programmers")
```
