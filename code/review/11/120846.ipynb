## 합성수 찾기


https://school.programmers.co.kr/learn/courses/30/lessons/120846

## 문제 풀이

```python
def solution(n):
    comp_cnt = 0
    for num in range(1,n+1):
        div_arr = [div for div in range(1,num+1) if num%div==0]
        if len(div_arr) >= 3:
            comp_cnt += 1
    return comp_cnt

solution(10)
```
