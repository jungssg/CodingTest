## 문제 풀이

```python
def solution(num_list, n):
    return num_list[n:]+num_list[:n]

solution([2, 1, 6],1)
```

```python
def solution(num_list, n):
    return sum([num_list[n:],num_list[:n]],[])

solution([2, 1, 6],1)
```
