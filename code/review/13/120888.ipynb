## 중복된 문자 제거


https://school.programmers.co.kr/learn/courses/30/lessons/120888

## 문제 풀이

```python
from collections import Counter

def solution(my_string):
    duplicate = [k for k,v in Counter(my_string).items() if v>1]
    for dup in duplicate:
        idx = my_string.find(dup)
        my_string = my_string[:idx+1]+my_string[idx+1:].replace(dup,'')
    return my_string

solution("We are the world"	)
```
