## 문제 풀이

```python
def solution(str_list):
    s = ''.join(str_list)
    i = dict(zip([s.find(x) for x in 'lr'], 'lr'))
    positve = [p for p in i if p>=0]
    if not positve:
        return []
    j = min(positve)
    if i[j]=='l':
        return list(s[:j])
    elif i[j]=='r':
        return list(s[j+1:])
    
solution(["u", "u", "l", "r"])
```
