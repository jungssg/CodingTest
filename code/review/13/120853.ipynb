## 문제 풀이

```python
def solution(s):
    s = s.split()[::-1]
    i = 0
    n = []
    while i<len(s):
        if s[i]=='Z':
            i+=2
            continue
        n.append(int(s[i]))
        i+=1
    return sum(n)

solution("10 Z 20 Z 1")
```
