## 문제 풀이

```python
def solution(price):
    if price >= 500000:
        return int(price*0.8)
    elif price >= 300000:
        return int(price*0.9)
    elif price >= 100000:
        return int(price*0.95)
    else:
        return price

solution(150000)
```

## 다른 풀이

```python
def solution(price):
    discounts = {500000: 0.8, 300000: 0.9, 100000: 0.95, 0: 1}
    for amount, rate in discounts.items():
        if price>=amount:
            return int(price*rate)

solution(150000)
```
