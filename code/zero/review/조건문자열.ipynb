## 문제 풀이

```python
def solution(ineq, eq, n, m):
    answer = 0
    
    case = ineq+eq
    if case == '<=':
        answer = n<=m
    elif case == '>=':
        answer = n>=m
    elif case == '<!':
        answer = n<m
    elif case == '>!':
        answer = n>m
    
    return int(answer)

solution("<","=",20,50)
```
