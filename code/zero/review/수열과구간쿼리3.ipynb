## 문제 풀이

```python
def solution(arr, queries):
    answer = []
    
    for i, j in queries:
        a, b = arr[i], arr[j]
        arr[i], arr[j] = b, a
    answer = arr
    
    return answer

solution([0, 1, 2, 3, 4], [[0, 3],[1, 2],[1, 4]])
```
