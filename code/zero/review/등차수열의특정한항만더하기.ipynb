## 문제 풀이

```python
def solution(a, d, included):
    answer = 0
    
    sequence = map(lambda x: a+d*x, range(len(included)))
    answer = sum([s for i, s in zip(included,sequence) if i])
    
    return answer

solution(3,4,[True, False, False, True, True])
```

## 다른 풀이

```python
def solution(a, d, included):
    answer = sum([a+d*i for i,b in enumerate(included) if b])
    return answer

solution(3,4,[True, False, False, True, True])
```
