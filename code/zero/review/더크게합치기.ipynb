## 문제 풀이

```python
def solution(a, b):
    answer = 0
    
    ab = int(''.join(map(str,[a,b])))
    ba = int(''.join(map(str,[b,a])))
    answer = max(ab, ba)
    
    return answer

solution(9,91)
```

## 다른 풀이

```python
def solution(a, b):
    answer = max(f'{a}{b}',f'{b}{a}')
    answer = int(answer)
    return answer
```
