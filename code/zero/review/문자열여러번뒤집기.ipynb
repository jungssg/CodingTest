## 문제 풀이

```python
def solution(my_string, queries):
    for s, e in queries:
        new = ''.join(reversed(my_string[s:e+1]))
        my_string = my_string[:s]+new+my_string[e+1:]
    return my_string

solution("rermgorpsam", [[2, 3], [0, 7], [5, 9], [6, 10]])
```

## 다른 풀이

```python
def solution(my_string, queries):
    answer = list(my_string)
    for s,e in queries:
        answer[s:e+1] = answer[s:e+1][::-1]
    return ''.join(answer)
```
