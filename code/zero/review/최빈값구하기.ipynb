## 문제 풀이

```python
def solution(array):
    answer = 0
    
    nums = list(set(array))
    cnts = []
    for num in nums:
        cnts.append(array.count(num))
    cnt = max(cnts)
    if cnts.count(cnt) > 1:
        answer = -1
    else:
        answer = nums[cnts.index(cnt)]
    
    return answer

solution([1, 2, 3, 3, 3, 4])
```

## 다른 풀이

```python
from collections import Counter

def solution(array):
    cnts = Counter(array).most_common(2)
    if cnts[0][0] != cnts[-1][0] and cnts[0][1] == cnts[-1][1]:
        return -1
    return cnts[0][0]

solution([1, 2, 3, 3, 3, 4])
```

```python
from collections import Counter

def solution(array):
    cnts = Counter(array)
    max_num = max(cnts, key=cnts.get)
    max_cnt = max(cnts.values())
    if list(cnts.values()).count(max_cnt) > 1:
        return -1
    else: 
        return max_num
    
solution([1, 2, 3, 3, 3, 4])
```
