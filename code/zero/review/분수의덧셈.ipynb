## 문제 풀이

```python
def solution(numer1, denom1, numer2, denom2):
    answer = []
    
    numer = numer1*denom2 + numer2*denom1
    denom = denom1*denom2
    
    div = []
    for n in range(1, denom+1):
        if denom%n==0:
            div.append(n)
    for n in reversed(div):
        if numer%n==0:
            answer = [numer//n, denom//n]
            break
    
    return answer

solution(9,2,1,3)
```

## 다른 풀이

```python
from fractions import Fraction
def solution(numer1, denom1, numer2, denom2):
    frac = Fraction(numer1, denom1) + Fraction(numer2, denom2)    
    return [frac.numerator, frac.denominator]

solution(9,2,1,3)
```

```python
import math
def solution(numer1, denom1, numer2, denom2):
    denom = math.lcm(denom1,denom2)
    numer = numer1*(denom//denom1) + numer2*(denom//denom2)
    return numer, denom

solution(9,2,1,3)
```
