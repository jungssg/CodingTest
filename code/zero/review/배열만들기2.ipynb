## 문제 풀이

```python
from itertools import product
def solution(l, r):
    limit = len(str(r))
    digit = {'0','5'}

    nums = [l, r]
    for comb in product(digit,repeat=limit):
        nums.append(int(''.join(map(str,comb))))
    nums = sorted(set(nums))

    start = nums.index(l)
    end = nums.index(r)

    if set(str(l))-digit:
        start += 1
    if set(str(r))-digit:
        end -= 1

    return nums[start:end+1] or [-1]

solution(10,20)
```

## 다른 풀이

```python
from itertools import product

def solution(l,r):
    digit = {'0','5'}
    limit = len(str(r))
    nums = [int(''.join(comb)) for comb in product(digit, repeat=limit)]
    return sorted([n for n in range(l,r+1) if n in nums]) or [-1]

solution(5,555)
```
