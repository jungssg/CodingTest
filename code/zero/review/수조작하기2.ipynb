## 문제 풀이

```python
def solution(numLog):
    answer = ''
    
    log = []
    for idx,num in enumerate(numLog):
        if idx == 0:
            continue
        log.append(num-numLog[idx-1])
    
    val = [+1,-1,+10,-10]
    key = ['w','s','d','a']
    for l in log:
        answer += key[val.index(l)]
    
    return answer

solution([0, 1, 0, 10, 0, 1, 0, 10, 0, -1, -2, -1])
```

## 다른 풀이

```python
def solution(numLog):
    valLog = []
    for idx,num in enumerate(numLog):
        if idx == 0:
            continue
        valLog.append(num-numLog[idx-1])
    
    key = dict(zip([+1,-1,+10,-10],'wsda'))
    return ''.join([key[val] for val in valLog])

solution([0, 1, 0, 10, 0, 1, 0, 10, 0, -1, -2, -1])
```
