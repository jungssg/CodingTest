## 문제 풀이

```python
def solution(number):
    return int(number)%9

solution("123")
```

```python
def solution(number):
    return sum([int(n) for n in number])%9

solution("123")
```

```python
def solution(number):
    return sum(map(int, number)) % 9

solution("123")
```
