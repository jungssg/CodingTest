## 문제 풀이

```python
def solution(n):
    answer = 0
    
    nums = range(1,n+1)
    even, odd = 0, 0
    for num in nums:
        if num%2==0:
            even+=num**2
        else:
            odd+=num
    answer = even if n%2==0 else odd
    
    return answer

solution(10)
```

## 다른 풀이

```python
def solution(n):
    answer = 0
    
    if n%2==0:
        answer = sum(map(lambda x:x**2, range(2,n+1,2)))
    else:
        answer = sum(range(1,n+1,2))
    
    return answer
```
