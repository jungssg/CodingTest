## 순서쌍의 개수

https://school.programmers.co.kr/learn/courses/30/lessons/120836

## 문제 풀이

```python
def solution(n):
    return len([(i, n//i) for i in range(1,n+1) if n%i == 0])

solution(100)
```

## 다른 풀이

```python
def solution(n):
    return len([i for i in range(1,n+1) if n % i ==0])

solution(100)
```

```python
def solution(n):
    cnt = 0
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            if i**2==n:
                cnt+=1
            else:
                cnt+=2
    return cnt

solution(100)
```
