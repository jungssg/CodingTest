## 외계행성의 나이

https://school.programmers.co.kr/learn/courses/30/lessons/120834

## 문제 풀이

```python
def solution(age):
    return ''.join([chr(int(n)+97) for n in str(age)])

solution(23)
```

## 다른 풀이

```python
age = 23
exo = str.maketrans('0123456789', 'abcdefghij')
# exo = dict(zip('0123456789','abcdefghij'))
str(age).translate(exo)
```

```python
str.maketrans('0123456789', 'abcdefghij')
```

```python
age = 23
exo = str.maketrans('0123456789', 'abcdefghij')
for n in str(age):
    print(chr(exo[ord(n)]))
```

```python
age = 23
exo = dict(zip(range(10),'abcdefghij'))
''.join([exo[int(n)] for n in str(age)])
```
