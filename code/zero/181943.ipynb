## 문자열 겹쳐쓰기

## 문제 풀이

```python
def solution(my_string, overwrite_string, s):
    answer = ''
    
    e = s + len(overwrite_string)
    answer += my_string[:s]
    answer += overwrite_string
    answer += my_string[e:]
    
    return answer

solution("He11oWor1d", "lloWorl", 2)
```

## 다른 풀이

```python
def solution(my_string, overwrite_string, s):
    answer = ''
    
    old = my_string[s:s+len(overwrite_string)]
    new = overwrite_string
    answer = my_string.replace(old, new)
    
    return answer

solution("He11oWor1d", "lloWorl", 2)
```

```python
def solution(my_string, overwrite_string, s):
    answer = ''
    
    old = my_string[s:s+len(overwrite_string)]
    new = overwrite_string
    answer = my_string[:s] + my_string[s:].replace(old, new, 1)
    
    return answer

solution("11oWor1He11oWor1d11oWor1", "lloWorl", 9)
```
