## 문제 풀이

```python
def solution(emergency):
    ordered = sorted(emergency[:])[::-1]
    return [ordered.index(e)+1 for e in emergency]

solution([3, 76, 24])
```

## 다른 풀이

```python
def solution(emergency):
    order = sorted(emergency, reverse=True)
    return [order.index(e)+1 for e in emergency]

solution([3, 76, 24])
```
