## 문제 풀이

```python
def solution(dot):
    x = int(dot[0]<0)
    y = int(dot[1]<0)
    
    if x == y:
        return 2*y+1
    else:
        return 2*(y+1)

solution([-7,9])
```

## 다른 풀이

```python
def solution(dot):
    quad = [(1,4),(2,3)]
    return quad[dot[0]<0][dot[1]<0]

solution([-7,9])
```
