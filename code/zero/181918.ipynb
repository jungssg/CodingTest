## 문제 풀이

```python
def solution(arr):
    i = 0
    stk = []
    while i < len(arr):
        if not stk:
            stk.append(arr[i])
            i+=1
            continue
        if stk[-1]<arr[i]:
            stk.append(arr[i])
            i+=1
        else:
            stk.pop()
    return stk

solution([1, 4, 2, 5, 3])
```
