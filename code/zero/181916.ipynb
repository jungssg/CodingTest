## 문제 풀이

```python
from collections import Counter

def solution(a, b, c, d):
    cnts = Counter([a,b,c,d])
    keys = cnts.keys()
    if len(keys) == 1:
        p = a
        return 1111*p
    if len(keys) == 2:
        p, q = keys
        if cnts[p]==cnts[q]:
            return (p+q)*abs(p-q)
        else:
            return (10*p+q)**2
    if len(keys) == 3:
        p, q, r = keys
        return q*r
    
    return min(keys)

solution(4,1,4,4)
```

```python
from collections import Counter

def solution(a, b, c, d):
    cnts = Counter([a,b,c,d])
    keys = list(cnts.keys())
    p = max(cnts, key=cnts.get)
    if len(keys) == 1:
        return 1111*p
    if len(keys) == 2:
        keys.remove(p)
        q = keys.pop()
        if cnts[p]==cnts[q]:
            return (p+q)*abs(p-q)
        else:
            return (10*p+q)**2
    if len(keys) == 3:
        keys.remove(p)
        q, r = keys
        return q*r
    else: 
        return min(keys)

solution(4,1,4,4)
```
