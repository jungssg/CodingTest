## 대소문자 바꿔서 출력하기


https://school.programmers.co.kr/learn/courses/30/lessons/181949

## 문제 풀이

```python
str = input()
out = ''
for c in str:
    upper_c = c.upper()
    lower_c = c.lower()
    out += (upper_c if c == lower_c else lower_c)
print(out)
```

## 다른 풀이

```python
str = input()
out = str.swapcase()
print(out)
```
