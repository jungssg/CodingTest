## 문제 풀이

```python
def solution(arr, queries):
    answer = []
    
    for s,e,k in queries:
        nums = [num for num in arr[s:e+1] if num > k]
        answer.append(-1 if len(nums)==0 else min(nums))
    
    return answer

solution([0, 1, 2, 4, 3],[[0, 4, 2],[0, 3, 2],[0, 2, 2]])
```
