## 문제 풀이

```python
def solution(num_list):
    answer = []
    
    last = num_list[-1]
    before = num_list[-2]
    
    if last > before:
        num_list.append(last-before)
    else:
        num_list.append(last*2)
        
    answer = num_list
    
    return answer

solution([2, 1, 6])
```
