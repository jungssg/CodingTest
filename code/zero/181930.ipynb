## 문제 풀이

```python
def solution(a, b, c):
    answer = 0
    
    case = len(set([a,b,c]))
    if case == 1:
        answer = (a + b + c) * (a**2 + b**2 + c**2 ) * (a**3 + b**3 + c**3 )
    elif case == 2:
        answer = (a + b + c) * (a**2 + b**2 + c**2 )
    elif case == 3:
        answer = a + b + c
    
    return answer

solution(2,6,1)
```
