## 원소들의 곱과 합


https://school.programmers.co.kr/learn/courses/30/lessons/181929

## 문제 풀이

```python
def solution(num_list):
    answer = 0
    
    reduce = 1
    for num in num_list:
        reduce*=num
    answer = int(reduce < sum(num_list)**2)
    
    return answer

solution([3, 4, 5, 2, 1])
```

## 다른 풀이

```python
from functools import reduce

num_list = [3, 4, 5, 2, 1]
int(reduce(lambda x,y:x*y, num_list) < sum(num_list)**2)
```
