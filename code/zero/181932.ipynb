## 문제 풀이

```python
def solution(code):
    answer = ''
    
    mode = 0
    for i, c in enumerate(code):
        if c == '1':
            mode = (mode+1)%2
            continue
        if mode == i%2:
            answer += c
    answer = answer if answer else 'EMPTY'
    
    return answer

solution("abcd1abc1abc")
```
