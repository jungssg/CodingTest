## 짝수 홀수 개수


https://school.programmers.co.kr/learn/courses/30/lessons/120824

## 문제 풀이

```python
def solution(num_list):
    answer = [0, 0]
    for num in num_list:
        if num%2==0:
            answer[0] += 1
        else:
            answer[1] += 1    
    return answer

solution([1, 3, 5, 7])
```

## 다른 풀이

```python
def solution(num_list):
    answer = [0,0]
    for n in num_list:
        answer[n%2]+=1
    return answer
```
