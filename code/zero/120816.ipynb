## 문제 풀이

```python
def solution(slice, n):
    answer = 0
    
    pizza = 1
    while pizza*slice < n:
        pizza += 1
    answer = pizza
    
    return answer

solution(7,10)
```

## 다른 풀이

```python
def solution(slice, n):
    if n % slice == 0:
        answer = n // slice
    else:
        answer = n // slice +1
    return answer

solution(7,10)
```

```python
def solution(slice, n):
    d, m = divmod(n, slice)
    # return d + int(m != 0)
    return d + int(bool(m))

solution(7,10)
```

```python
def solution(slice, n):
    return ((n - 1) // slice) + 1

solution(7,10)
```

```python
import math
def solution(slice, n):
    return math.ceil(n/slice)

solution(7,10)
```
