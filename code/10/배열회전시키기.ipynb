## 문제 풀이

```python
def solution(numbers, direction):
    if direction == "right":
        numbers.insert(0,numbers[-1])
        return numbers[:-1]
    if direction == "left":
        numbers.append(numbers[0])
        return numbers[1:]

solution([1,2,3],"right")
```
